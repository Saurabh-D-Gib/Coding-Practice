# Instance Methods

```python
## Define a class with Instance Methods
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name=name
        self.age=age
    #Instance methods
    def bark(self):
        print(f"{self.name} says woof")

dog1=Dog("Buddy",3)
dog1.bark()
dog2=Dog("Lucy", 4)
dog2.bark()
```

# Modelling a bank account

```python
# Define a class for bank acocunt
class BankAccount:
    # Make Instance Variable using constrctor
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    # make instance method
    def deposit(self, amount):
        self.balance+=amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount>self.balance:
            print("Insufficient funds!")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New Balance is {self.balance}")
    
    def get_balance(self):
        return self.balance
    
## Create an Object for an account
account=BankAccount("Krish", 5000)
print(account.balance)
```

Student Class
Task: Create a class to store student details.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)

s1 = Student("Rahul", 101, 85)
s1.display()
```

Rectangle Class

Task: Calculate area and perimeter.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(10, 5)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

Class with Static Method

Task: Check if number is prime.

```python
class MathUtils:
    @staticmethod
    def is_prime(n):
        if n <= 1:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

print(MathUtils.is_prime(11))
```

Counter Class

Task: Increment & reset counter.

```python
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def reset(self):
        self.count = 0

c = Counter()
c.increment()
c.increment()
print(c.count)
c.reset()
print(c.count)
```

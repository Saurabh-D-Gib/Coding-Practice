```python
class Employee:
    # constructor
    def __init__(self, name, age, dept, role):
        self.name=name
        self.age=age
        self.dept=dept
        self.role=role
emp1=Employee("Saurabh Kumar Dewnagan", 24, "CBSE", "Coordinator")
print(emp1.name)
```

```python
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name=name
        self.age=age
    # Instance methods
    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog("Buddy", 3)
dog1.bark()

dog2=Dog("Lucy", 4)
dog2.bark()
```

```python
class BankAccount:
    # Make instance variable using constructor
    def __init__(self,owner, balance=0):
        self.owner=owner
        self.balance=balance
    # Make instance method 
    def deposit(self,amount):
        self.amount+=amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdrawn(self, amount):
        if amount>self.balance:
            print("Insufficient funds!")
        else:
            self.balance-=amount
            print(f"{amount}is withdrawn. New Balance is {self.balance}")

    def get_balance(self):
        return self.balance
    
## Create an object for an account
account=BankAccount("Saurabh", 5000)
print(account.balance)
```

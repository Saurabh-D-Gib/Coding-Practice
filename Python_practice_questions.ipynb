```python
# Find Prime Number
n = int(input())

if n <=1:
    print("Not a prime number")
else:
   for i in range(2,n):
      if n % 2 == 0:
         print("not a Prime Number")
         break
      else:
         print("Prime Number")
      
```

```python
# Find the maximum integer from an array of integers
n=int(input())
arr=[]
for i in range(n):
    arr.append(int(input()))
max=arr[0]
for i in range(1,n):
    if arr[i] > max:
        max=arr[i]
print("Largest integer in given array is", max)
```
